## Dealing with the missing values
- missing value are either --> n/a,NaN,0 or blank number
#### Method
1. try to recover the missing Value and Recheck for the mistake
2. delete the column if it did'nt effect on the data
3. Replace(how)
   - Taking the average and replace it with the NaN cell
   - frequency or Mode replacement
   - ML algo applied on it
   - replaced based on other functions
   - leave it like that 

```python
# checking first the missing Value
ship.isnull().sum()
```

```python
# checking columns and rows
ship.shape
```

```python
# removing the missing values in specific columns
ship.dropna(subset=['deck'],axis=0,inplace=True)
```

```python
# checking again
ship.isnull().sum()
```

```python
# to drop all missing values
ship = ship.dropna()
ship.isnull().sum()
```

```python
# checking columns again
ship.shape
```

### Replacing the missing Values with mean method

```python
ship2.isnull().sum()
```

```python
# first getting the mean of that column that have missing Values
mean_age = ship2['age'].mean()
mean_age
```

```python
# replacing the missing the Values with mean
ship2['age'] = ship2['age'].replace(np.nan,mean_age)
# checking the age column missing Values
ship2['age'].isnull().sum()
```

```python
# thus the deck is categorical variable so we can't cal their mean so we are calculating their mode for replacement
mode_deck = ship2['deck'].mode()[0]
# replacing the mode with missing value
ship2['deck'] = ship2['deck'].fillna(mode_deck)
#checking 
ship2.isnull().sum()
```

```python
ship2.dropna().isnull().sum()
```

## Normalize the data
- Uniform the data
- they have same impact
- also for computational purposes
- make data in 0,1 range (assss passs le aty hain ta ky compare kr saky)

```python
ship_normal=ship[['age','fare']]
ship_normal.head()
```

## Method of normalize
- simple feature scalling
    - x(new)=x(old)/x(max)
- min-max method
- Z-score(standard score)-3 to +3
- log transformation

```python
# simple feature scalling
ship_normal['fare'] = ship_normal['fare']/ship_normal['fare'].max()
ship_normal.head()
```

```python
# min - max method
ship_normal['fare'] = (ship_normal['fare']- ship_normal['fare'].min())/(ship_normal['fare'].max()-ship_normal['fare'].min())
ship_normal.head()
```

```python
# Z score method
ship_normal['fare'] = (ship_normal['fare']-ship_normal['fare'].mean())/ship_normal['fare'].std()
ship_normal.head()
```

```python
ship3.head()
```

```python
# log transformmation
ship3['fare'] = np.log(ship3['fare'])
ship3.head()
```

# Binning
- Grouping the values into smaller group(bins)
- convert numeric into categorical(male,female,kids)
- to have better understanding of groups
    - low vs mid vs high price

```python
# using the bin method
bins = np.linspace(min(ship3['age']),max(ship3['age']),4)
age_grps = ["kids","young","old"]
ship3['age_grps'] = pd.cut(ship3['age'],bins,labels = age_grps,include_lowest = True )
ship3[['age','age_grps']].head()
```
