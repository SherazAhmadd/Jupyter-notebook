### customization in boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pf
ship=sns.load_dataset("titanic")
ship.describe()
# drawing plot
sns.boxplot(x="survived",y="age",hue="sex",data=ship,
           showmeans=True,meanprops={"marker":"+",
                                     "markersize":"10",
                                      "markeredgecolor":"white" })
plt.xlabel("how many survived"),
plt.ylabel("age(years)"),
plt.title("mera graph",size=10,weight='bold')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pf
ship=sns.load_dataset("titanic")
ship.describe()
# drawing plot
g=sns.FacetGrid(ship,row="class",col="sex")
g.map(sns.scatterplot, "age", "fare")
plt.show()
```

```python
import seaborn as sns
sns.set_theme(style="ticks")

dots = sns.load_dataset("dots")

# Define the palette as a list to specify exact values
palette = sns.color_palette("rocket_r")

# Plot the lines on two facets
sns.relplot(
    data=dots,
    x="time", y="firing_rate",
    hue="coherence", size="choice", col="align",
    kind="line", size_order=["T1", "T2"], palette=palette,
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)
```
