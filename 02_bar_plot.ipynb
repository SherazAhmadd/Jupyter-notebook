```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
# load data set
phool = sns.load_dataset("iris")
phool
sns.set_style("dark")
# draw the barplot
sns.barplot(x="species",y="sepal_width",data=phool)
plt.title("iris flower barplot")
plt.show()
```
