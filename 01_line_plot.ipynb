```python
import seaborn as sns
import matplotlib.pyplot as plt

# load datasets
flower = sns.load_dataset("iris")
flower
# draw lineplot
sns.lineplot(x="sepal_length",y="sepal_width",data= flower)
plt.title("iris flowers lineplot")
plt.show()
```

# adding limits of graph

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load datasets
flower = sns.load_dataset("iris")
flower
# draw lineplot
sns.lineplot(x="sepal_length",y="sepal_width",data= flower)
plt.title("iris flowers lineplot")
# adding the x and y limits
plt.xlim(4)
plt.ylim(2)
plt.show()
```

# set styles
#### (layout appearence)
- darkgrids
- whitegrids
- darks
- whites
- ticks

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

# load datasets
flower = sns.load_dataset("iris")
flower

sns.set_style("darkgrid")

# draw lineplot
sns.lineplot(x="sepal_length",y="sepal_width",data= flower)
plt.title("iris flowers lineplot")
# displaying
plt.show()
```
